```python
# 1
def minimumDeleteSum(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n+1) for _ in range(m+1)]

    for i in range(m-1, -1, -1):
        dp[i][n] = dp[i+1][n] + ord(s1[i])

    for j in range(n-1, -1, -1):
        dp[m][j] = dp[m][j+1] + ord(s2[j])

    for i in range(m-1, -1, -1):
        for j in range(n-1, -1, -1):
            if s1[i] == s2[j]:
                dp[i][j] = dp[i+1][j+1]
            else:
                dp[i][j] = min(dp[i+1][j] + ord(s1[i]), dp[i][j+1] + ord(s2[j]))

    return dp[0][0]
```

```python
# 2
def checkValidString(s):
    paren_stack = []

    for c in s:
        if c == '(' or c == '*':
            paren_stack.append(c)
        elif c == ')':
            if paren_stack and (paren_stack[-1] == '(' or paren_stack[-1] == '*'):
                paren_stack.pop()
            else:
                return False

    while paren_stack:
        if paren_stack[-1] == '(' or paren_stack[-1] == '*':
            paren_stack.pop()
        else:
            break

    return len(paren_stack) == 0
```

```python
#3
def minDistance(word1, word2):
    m, n = len(word1), len(word2)
    dp = [[0] * (n+1) for _ in range(m+1)]

    for i in range(m+1):
        dp[i][0] = i

    for j in range(n+1):
        dp[0][j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):
            if word1[i-1] == word2[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = min(dp[i-1][j] + 1, dp[i][j-1] + 1)

    return dp[m][n]
```

```python
#4
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def buildTree(s):
    if not s:
        return None

    i = s.find('(')

    if i != 0:
        root_val = int(s[:i])
        root = TreeNode(root_val)
    else:
        root = None

    j = find_matching_parenthesis(s, i)

    left_substring = s[i+1:j]
    root.left = buildTree(left_substring)

    if j+1 < len(s):
        right_substring = s[j+2:-1]
        root.right = buildTree(right_substring)

    return root

def find_matching_parenthesis(s, start):
    count = 0

    for i in range(start, len(s)):
        if s[i] == '(':
            count += 1
        elif s[i] == ')':
            count -= 1

        if count == 0:
            return i

    return -1
s = "4(2(3)(1))(6(5))"
root = buildTree(s)
```

```python
# 5
def compress(chars):
    write = 0
    count = 1

    for read in range(1, len(chars) + 1):
        if read < len(chars) and chars[read] == chars[read - 1]:
            count += 1
        else:
            chars[write] = chars[read - 1]
            write += 1
            if count > 1:
                count_str = str(count)
                for digit in count_str:
                    chars[write] = digit
                    write += 1
            count = 1

    return write
```

```python
#6
from collections import defaultdict

def findAnagrams(s, p):
    p_freq = defaultdict(int)
    window_freq = defaultdict(int)
    result = []

    # Initialize p_freq with frequencies of characters in p
    for char in p:
        p_freq[char] += 1

    left = right = 0
    window_size = len(p)
    s_len = len(s)

    while right < s_len:
        if s[right] in p_freq and window_freq[s[right]] < p_freq[s[right]]:
            window_freq[s[right]] += 1
            right += 1

            if right - left == window_size:
                if window_freq == p_freq:
                    result.append(left)

                window_freq[s[left]] -= 1
                left += 1
        else:
            if s[right] not in p_freq:
                window_freq.clear()
            else:
                window_freq[s[right]] -= 1

            right += 1
            left = right

    return result
```

```python
#7
def decodeString(s):
    stack = []

    for char in s:
        if char != ']':
            stack.append(char)
        else:
            # Decode the substring inside the square brackets
            substr = ''
            while stack and stack[-1] != '[':
                substr = stack.pop() + substr

            # Pop the opening square bracket '['
            stack.pop()

            # Find the repeat count 'k'
            k = ''
            while stack and stack[-1].isdigit():
                k = stack.pop() + k
            k = int(k)

            # Repeat the substring 'k' times and push it onto the stack
            stack.append(substr * k)

    return ''.join(stack)
```

```python
#8
def buddyStrings(s, goal):
    if len(s) != len(goal):
        return False

    diff_s = []
    diff_goal = []

    for i in range(len(s)):
        if s[i] != goal[i]:
            diff_s.append(i)
            diff_goal.append(i)

    if len(diff_s) == 0:
        # s and goal are already the same
        return len(set(s)) < len(s)

    if len(diff_s) != 2 or len(diff_goal) != 2:
        return False

    i, j = diff_s[0], diff_s[1]
    if s[i] == goal[j] and s[j] == goal[i]:
        return True

    return False
```
